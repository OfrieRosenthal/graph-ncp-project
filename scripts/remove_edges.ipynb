```python
import networkx as nx

G = nx.read_edgelist("../datasets/facebook/facebook_combined.txt", nodetype=int)
print("Before:", G.number_of_edges(), "edges")

# Remove self-loops
G.remove_edges_from(nx.selfloop_edges(G))
print("After removing self-loops:", G.number_of_edges(), "edges")

# Make undirected & deduplicate
G = nx.Graph(G)

print("After making undirected & deduplicating:", G.number_of_edges(), "edges (cleaned)")
nx.write_edgelist(G, "../datasets/facebook/facebook_combined_clean.txt", data=False)
```

```python
import networkx as nx


loops = list(nx.selfloop_edges(G))
print(f"Found {len(loops)} self-loops:")
for u, v in loops:
    print(u, v)
```

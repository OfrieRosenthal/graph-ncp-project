```python
csv_file = "/home/ofrie.r/graph-ncp-project/datasets/facebook_pages/musae_facebook_edges.csv"  # input CSV
txt_file = "/home/ofrie.r/graph-ncp-project/datasets/facebook_pages/musae_facebook_edges.txt"  # output TXT

with open(csv_file, 'r') as fin, open(txt_file, 'w') as fout:
    next(fin)  # skip the header line
    for line in fin:
        # remove newline and replace comma with space
        fout.write(line.replace(',', ' '))
```

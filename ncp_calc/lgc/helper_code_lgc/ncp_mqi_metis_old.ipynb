```python
import sys
sys.path.append("../")

from localgraphclustering.GraphLocal import GraphLocal
from localgraphclustering.ncp import NCPData
from localgraphclustering.ncpplots import NCPPlots
import matplotlib.pyplot as plt
```

```python

# 1. Load graph
from localgraphclustering.GraphLocal import GraphLocal

G = GraphLocal("../facebook_combined.txt", separator=r"\s+")

# 2. Run NCP with MQI (Metis+MQI style)
ncp = NCPData(G)
ncp.mqi(ratio=0.3, nthreads=4)

# 3. Plot NCP
plotter = NCPPlots(ncp)
fig, ax, min_tuples = plotter.cond_by_vol(nbinsx=100, log=True)

plt.show()
```

```python
import networkx as nx
G_nx = nx.read_edgelist("../facebook_combined.txt", delimiter=" ", nodetype=int)

nx.number_connected_components(G_nx)
```
